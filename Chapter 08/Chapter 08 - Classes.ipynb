8.3 Encapsulation and abstraction
---

Description

---
Encapsulation involves the bundling of data (attributes) and method (functions) that operate on data within a single unit. A single unit is known as a class. It allows one to access and modify the attributes while keeping the attributes private.

Syntax

---
Incapsulation

      class ClassName:
          def _ _init_ _(self):
              self.__attribute = value    # Private attribute

          def get_attribute(self):
              return self.__attribute

          def set_attribute(self, value):
              self.__attribute = value

Description

---

Abstraction is a way to simplify the complex reality by creating a class and hide some unnecessary details. You can implement abstraction using classes and methods, where you show only the relevant details and hiding the internal implementation.

Syntax

---
Abstraction

    class ClassName:
        def public_method(self):
            # Interface exposed to the user
            self.__internal_method()

        def __internal_method(self):
            # Hidden implementation details
            pass

Examples

---

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance      # Encapsulation

    def deposit(self, amount):        # Abstraction
        self.__balance += amount

    def get_balance(self):
        return self.__balance
```

Note

---

**_ _balance** demonstrates encapsulation by making the attribute private. **deposit()** and **get_balance()** demonstrate abstraction by providing a simple interface to interact with the account without exposing internal implementation details.

8.4 Private property
---

Description

---
A **public attribute** is intended to be accessed and modified from anywhere in the program. It has **no leading underscores** and is part of the class's public interface.

An attribute prefixed with a **single underscore (_)** is conventionally considered **protected or internal**, indicating that it should only be accessed within the class or its subclasses. Python does not enforce this restriction; the underscore simply serves as a signal to developers that the attribute is intended for internal use.

An attribute prefixed with a **double underscore (__)** triggers name mangling and is conventionally considered **private**, making it more difficult to access or modify directly from outside the class. However, Python does not provide true private attributes, and such attributes can still be accessed if their mangled names are known.

Syntax

---
             public_variable

            _protected_variable

          _ _private_variable

 The **_ _ kind** attribute is intended **to be private**. The **double underscore ( _ _ )** triggers Python's name-mangling mechanism, making accidental access or modification more difficult. Assigning a value to **myTrain.__kind** does not modify the original private attribute; instead, it creates a new attribute. Although this provides a degree of protection, the original attribute can still be accessed if its mangled name is known.

 Although **_length is intended to be protected**, Python still allows direct access to it.  The **speed attribute is public** and can be accessed or modified directly from anywhere in the program.
```python
class Train:

    # Constructor
    def __init__(self, kind, length, speed):

        # Making the kind property private
        self.__kind      = kind

        # Making the length property protected
        self._length     = length

        # Making the speed property public
        self.speed       = speed

    # Adding methods of the class
    def speed_up(self):
        self.speed += 1

    def speed_down(self):
        self.speed -= 1

    def show_speed(self):
        print("The current train is ", self.__kind, self._length, "miles long, having speed of ", self.speed)

# Creating an object
myTrain = Train("High-speed", 2, 0)

# Accessing the speed and kind properties directly
myTrain.__kind    = "Regional"
myTrain._length   = 23
myTrain.speed     = -200

myTrain.show_speed()
print(myTrain._Train__kind) #accesing the private attribute
```

8.5 Private method
---

Description

---
**Public**, **protected**, and **private methods** are used to indicate the intended level of access to class methods by following naming conventions rather than strict access enforcement.

A **public method**, which has **no leading underscores** (e.g., **method()**), is intended to be accessed from anywhere inside or outside the class. Public methods form part of the class's public interface and are designed for use by other objects and programs.

A **protected method**, marked with a single leading underscore (e.g., **_method()**), indicates that the method is intended for use only within the class and its subclasses. Although it can still be accessed from outside the class, doing so violates the convention and should generally be avoided.

A **private method**, denoted by two leading underscores (e.g., **_ _ method()**), undergoes **name mangling**. Python internally renames the method to include the class name (e.g., _ClassName _ _ method), making it more difficult to access from outside the class. While this does not provide true privacy, it offers a stronger indication that the method is intended for internal use only and helps prevent accidental access or overriding.

Syntax

---
             myPublicMethod()

            _myProtectedMethod()
            
          _ _myPrivateMethod()

Examples

---

```python
class Train:

    # Constructor
    def __init__(self, kind, length, speed):

        # Making the kind property private
        self.__kind = kind

        # Making the length property protected
        self._length = length

        # Making the speed property public
        self.speed = speed

    # Public method
    def show_train(self):
        print("The current train is", self.__kind,
              self._length, "miles long, having a speed of",
              self.speed, "miles per hour")

    # Protected method
    def _check_length(self):
        print("Checking train length...")

    # Private method
    def __check_kind(self):
        print("Checking train type...")


# Creating an object
myTrain = Train("High-speed", 2, 0)

# Accessing methods directly
myTrain.show_train()      # Public method

myTrain._check_length()   # Protected method

# This creates an error because the method is private
# myTrain.__check_kind()

# Accessing the private method through name mangling
myTrain._Train__check_kind()
```

8.6 Class Variable
---

Description

---
A **class variable** is a variable that belongs to the class itself rather than to individual objects of the class. Unlike instance variables, which are created separately for each object using self, a class variable is shared by all instances of the class.

Class variables are typically used to store information that is common to all objects of a class, such as constants, counters, or default values.


Syntax

---

          class ClassName:
                class_variable = value

Examples

---

```python
class User:
    count = 0  # Class variable

    def __init__(self, name):
        self.name = name
        User.count += 1

# Create objects
user1 = User("Alice")
user2 = User("Bob")
user3 = User("Charlie")

print("Number of users:", User.count)
```

```python
class Train:

    # Class variable
    company = "National Railway"

    def __init__(self, kind, speed):
        self.kind = kind      # Instance variable
        self.speed = speed    # Instance variable


# Creating objects
train1 = Train("Passenger", 60)
train2 = Train("Freight", 45)

# Accessing the class variable
print(train1.company)
print(train2.company)
```

8.7 Enheritance
---

Description

---
**Inheritance** is a fundamental concept in object-oriented programming (OOP) that **allows one class, known as a child class or subclass, to inherit attributes and methods from another class, known as a parent class or superclass**. Inheritance promotes code reuse, modularity, and extensibility by enabling developers to build new classes based on existing ones.

In Python, **inheritance** is implemented by specifying the parent class in parentheses when defining the child class (e.g., **class Child(Parent)**:). The child class automatically gains access to the attributes and methods of the parent class. It can use these inherited features as they are, override them to provide specialized behavior, or extend them by adding new functionality.

In addition to inheriting attributes and methods, **a child class can define its own properties and methods that are specific to its purpose**. Python also supports multiple inheritance, which allows a class to inherit from more than one parent class. This enables a child class to combine functionality from multiple sources, providing greater flexibility in class design.

Syntax

---
          Parent class

            class ParentClass:
                def __init__(self):
                  print("Parent constructor")

                def parentMethod(self):
                  print("This is a method in the parent class")

          Child class inheriting from ParentClass

            class ChildClass(ParentClass):
                def __init__(self):
                  super().__init__()  # Call the parent class constructor
                  print("Child constructor")

                def childMethod(self):
                  print("This is a method in the child class")

          Creating an object of the child class

          myObject = ChildClass()
          myObject.parentMethod()   # Inherited from ParentClass
          myObject.childMethod()    # Defined in ChildClass


Examples

---

```python
# Define a parent or superclass class.
# Animal is the broad category, encompassing all living organisms.

class Animal:
    pass

# Define a child or subclass class that inherits from Animal.
# Mammals are a specific class of animals, they are characterized
# by certain distinctive features like mammary glands and hair.

class Mammal(Animal):
    pass

# Define a child or subclass class that inherits from Mammal.
# Dog is the specific specie within the mammal class.

class Dog(Mammal):
    pass

#issubclass(Mammal,Animal)
#issubclass(Dog,Mammal)
#issubclass(Dog,Animal)

myDog = Dog()
print(myDog.__dict__)
```

Note

---
The super() function is used to call methods from a parent class (or superclass) within a child class (or subclass). It is most commonly used to invoke the parent class's constructor (__init__()) so that the child class can inherit and initialize attributes defined in the parent class.

The super() function can also be used to extend the functionality of inherited methods without completely overriding them. This allows a child class to reuse the behavior of the parent class while adding its own specialized functionality.

When using super(), you do not need to explicitly pass the self parameter. Python automatically binds the method call to the current object instance, making the syntax simpler and less error-prone.

Syntax

---
          class ParentClass:
              def __init__(self):
              print("Parent constructor")
        
          class ChildClass(ParentClass):
              def __init__(self):
              super().__init__()  # Calls Parent's __init__
              print("Child constructor")

Examples

---

```python
# Define a parent or superclass class.
# Animal is the broad category, encompassing all living organisms.

class Animal:
    message = "This message is from Animal superclass"
    def __init__(self,name):
      self.name = name

# Define a child or subclass class that inherits from Animal.
# Mammals are a specific class of animals, they are characterized
# by certain distinctive features like mammary glands and hair.

class Mammal(Animal):
    def __init__(self, name, color):
      super().__init__(name)
      self.color = color
      print(super().message)

# If you do not specify the constructor explicitely in Dog subclass
# Python will look for it in superclass

class Dog(Mammal):
    pass

myDog = Dog("Rex", "golden")
```

8.8 Method resolution order
---

Description

---
Method Resolution Order (MRO) is the order in which Python looks for a method or attribute when it is called on an instance of a class that involves inheritance. It becomes especially important in cases of multiple inheritance. When a class inherits from multiple paent classes, Python uses C3 linearization algorithm, also known as C3 MRO, to determine the order in which classes are searched.

Syntax

---
          ClassName.mro()
          
          ClassName._ _ mro _ _                 

Examples

---

```python
class Animal:
    pass

class Mammal(Animal):
    pass

class Canine(Mammal):
    pass

class Dog(Canine):
    pass

print(Dog.mro())   # returns a list

print(Dog.__mro__) # returns a tuple
```

8.9 Diamond problem
---

Description

---
There is a specific issue associated with multiple inheritance known as the **Diamond Problem**. It occurs when a class inherits from two parent classes that both inherit from the same superclass, creating a diamond-shaped inheritance hierarchy.

In such a situation, ambiguity can arise in the **Method Resolution Order (MRO)**, which determines the order in which Python searches for methods and attributes. If the same method is inherited through multiple paths, it may be unclear which implementation should be used when the method is called.

Python resolves the Diamond Problem using the Method Resolution Order (MRO) and the **C3 Linearization Algorithm**, which establish a consistent and predictable order for method lookup. This ensures that each class in the inheritance hierarchy is visited only once and eliminates ambiguity when inherited methods are invoked.

Syntax

---
          class A:
                 pass

          class B(A):
                 pass

          class C(A):
                 pass

          class D(B, C):
                 pass

<img src="https://github.com/HandsOnPythonProgramming/Hands-On-Python-Programming/blob/main/Images/dimond_problem.jpg?raw=true" width=750 />

Examples

---

```python
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Mammal(Animal):
    def speak(self):
        print("Mammal makes a sound")

class Reptile(Animal):
    def speak(self):
        print("Reptile makes a sound")

class Dog(Mammal, Reptile):
    pass

myDog = Dog()
myDog.speak()

print(Dog.mro())
```

```python
class Mammal:
    def speak(self):
        print("Mammal makes a sound")

class Feline(Mammal):
    def speak(self):
        print("Feline makes a sound: Meow, meow!")

class Canine(Mammal):
    def speak(self):
        print("Canine makes a sound: Woof, woof!")

class Dog(Canine, Mammal):
    pass

# Create instances of Dog
myDog = Dog()

# Call the speak method on both
myDog.speak()  # Which `speak()` will be called?

# Let's check MRO
print(Dog.mro())
```

 8.10  Method overriding
---

Description

---
Method overriding is an object-oriented programming feature that allows a subclass to redefine a method inherited from its superclass. The subclass creates a method with the same name as the method in the parent class, replacing the inherited version. This allows the subclass to provide behavior that is more appropriate for that specific class. When the method is called on an object of the subclass, Python executes the overridden method defined in the subclass rather than the method in the superclass.

Syntax

---
        class ParentClass:
          def method_name(self):
              # Parent class implementation
        pass

        class ChildClass(ParentClass):
          def method_name(self):
              # Overridden implementation
        pass

Note

---

Method overriding occurs when a subclass defines a method with the same name as a method in its superclass. The method in the subclass replaces the inherited version, allowing the subclass to provide its own implementation. When the method is called on an instance of the subclass, the overridden method in the subclass is executed instead of the method defined in the superclass. This enables subclasses to customize or extend the behavior inherited from their parent classes.

Examples

---

```python
# Define a parent or superclass class.
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        pass  # This method will be overridden in subclasses

# Define a child or subclass class that inherits from Animal.
class Dog(Animal):
    def speak(self):
        print ("Woof, Woof!")

# Define a child or subclass class that inherits from Animal.
class Cat(Animal):
    def speak(self):
        print ("Meow, Meow!")

# Create instances of the subclasses
myDog = Dog("Rex")
myCat = Cat("Misty")

# Call the speak method on each instance
myDog.speak()
myCat.speak()
```

8.11 Polymorphism
---

Description

---
Polymorphism is an object-oriented programming concept that allows the same method to be used with different types of objects. Although the method call may be identical, each object can respond in its own way by providing its own implementation of the method. This enables objects of different classes to be treated through a common interface while exhibiting class-specific behavior.

Polymorphism is commonly achieved through method overriding, where a subclass provides its own implementation of a method inherited from a superclass. As a result, the same method call can produce different outcomes depending on the type of object that invokes it. This flexibility promotes code reusability, extensibility, and maintainability.

In simple terms, polymorphism is the ability of a single method call to be executed on different objects, with each object responding according to its own implementation.

Syntax

---
          class ParentClass:
                def method_name(self):
                    pass

          class ChildClass1(ParentClass):
                def method_name(self):
                    # Implementation for ChildClass1
                    pass

          class ChildClass2(ParentClass):
                def method_name(self):
                    # Implementation for ChildClass2
                    pass

          def function_name(object):
              object.method_name()

          object1 = ChildClass1()
          object2 = ChildClass2()

          function_name(object1)
          function_name(object2)

Examples

---

```python
# Define a parent or superclass class.
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        pass  # This method will be overridden in subclasses

# Define a child or subclass class that inherits from Animal.
class Dog(Animal):
    def speak(self):
        print ("Woof, Woof!")

# Define a child or subclass class that inherits from Animal.
class Cat(Animal):
    def speak(self):
        print ("Meow, Meow!")

# Create instances of Dog and Cat
myDog = Dog("Rex")
myCat = Cat("Misty")

# Create a function that takes an Animal object and makes it speak
# When myDog and myCat instances are passed to the function,
# it is showing polymorphism. The function works with objects of
# different classes and calls the appropriate speak method for
#each object.

def animalSpeak(animal):
    return animal.speak()

# Call the make_animal_speak function with different animal objects
animalSpeak(myDog)
animalSpeak(myCat)
```

Note

---

This example demonstrates polymorphism in Python. The Animal class defines a common interface through the speak() method, which is overridden by the Dog and Cat subclasses. Each subclass provides its own implementation of the speak() method.

The animalSpeak() function accepts an object of type Animal. However, because Dog and Cat are subclasses of Animal, objects of either class can be passed to the function. When the speak() method is called, Python automatically executes the version of the method associated with the object's actual class.

As a result, the same method call, animal.speak(), produces different behaviors depending on whether the object is a Dog or a Cat. This ability of a single interface to work with objects of different types is known as polymorphism.
