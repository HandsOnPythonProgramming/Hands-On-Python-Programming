1.2 Comments
---

Description

---
Comments are an essential part of programming, as they make code easier to read and maintain by explaining its purpose and logic. In Python, single-line comments start with the # symbol, and everything after it is ignored by the interpreter. For longer explanations, you can use triple single quotes (' ' ') or triple double quotes (" " "), which are technically string literals but are often used as multi-line comments or docstrings (documentation strings) inside functions, classes, and modules.

Note

---

A docstring is a type of string in Python language that is used to document a module, class, or function. It provides documentation describing the purpose, behavior, and usage of the code. Docstrings are not executed as part of the program.
The docstring can be accessed via the **help()** function or the .__ _doc_ __ attribute.

Examples

---

The Student class demonstrates how to use multi-line comments (docstrings) in Python to document the class, its attributes, and its methods. In this example, the class has a docstring that clearly outlines the class name, attributes (the student's name), and methods (the greet() method, which prints a personalized greeting). The docstring is placed at the beginning of the class and serves as documentation, making it easier for others to understand the class’s functionality.

```python
class Student:
    '''
    Class: Student
    Description:
        A simple class representing a student with a name.
        The class provides a method to greet the student.

    Attributes:
        student_name (str): The name of the student.

    Methods:
        __init__(self, student_name): Initializes a new Student object with a given name.
        greet_student(self): Prints a personalized greeting using the student's name.

    Example:
        student = Student("Alice")
        student.greet_student()  # Output: Hello, Alice!
    '''

    def __init__(self, student_name):
        self.student_name = student_name

    def greet_student(self):
        print(f"Hello, {self.student_name}!")
```

This description highlights how the multi-line comment (docstring) is used to document the class, making it easier for other developers to understand its purpose. The ___init_ __() constructor and greet() method are also described to show how the class interacts with its attributes and methods.

Calling help(Student) will display the class documentation in an interactive format, which includes the docstring.

```python
help(Student)
```

You can also access the docstring directly by printing Student.__ _doc_ __

```python
print(Student.__doc__)
```

It helps provide important context to anyone reading or maintaining the code, such as collaborators, or even yourself when you revisit your code after a period of time.

1.3 Variables
---

Description

---
Variables are used to store data values that can be referenced and manipulated later in a program. In Python, a variable is created when you assign it a value using the assignment operator (=), and unlike many other programming languages, you do not need to explicitly declare its type. Python automatically determines the type based on the value assigned. A variable’s name serves as a reference to an object in memory, and each object can belong to different data types, holding corresponding values.

Syntax

---
          variable_name = value
