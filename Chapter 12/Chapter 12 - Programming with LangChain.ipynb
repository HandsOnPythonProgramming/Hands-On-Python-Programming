# 12.1 Understanding Large Language Models (LLMs)

Large Language Models (LLMs) are advanced artificial intelligence systems trained on massive amounts of text data to understand, generate, summarize, and analyze human language. Modern LLMs use a deep learning architecture called the Transformer, which allows the model to process relationships between words and concepts efficiently. The workflow of an LLM typically includes three major stages: Tokenization, where text is broken into smaller units called tokens; Embedding (Vectorization), where tokens are converted into numerical vector representations; and the Transformer Neural Network, which learns contextual relationships and generates intelligent responses.

**Tokenization** converts human language into smaller units called tokens that the model can process mathematically.

<img src="https://github.com/HandsOnPythonProgramming/Hands-On-Python-Programming/blob/main/Images/llm_tokenization_diagram.png?raw=true" width=550 />

**Embedding / Vectorization**

After tokenization, words are transformed into high-dimensional numerical vectors called embeddings. These vectors capture semantic meaning and relationships between words.

<img src="https://github.com/HandsOnPythonProgramming/Hands-On-Python-Programming/blob/main/Images/llm_vectorization_diagram.png?raw=true" width=650 />

# 12.2 LangChain Framework

---

Description

---

**LangChain** is an open-source framework designed to simplify the development of applications powered by large language models (LLMs). It provides modular building blocks for chaining together prompts, tools, memory, and external data sources, allowing developers to create more complex and context-aware AI systems. Instead of interacting with an LLM in isolation, **LangChain** enables workflows where models can retrieve information from databases, call APIs, and maintain conversational state over time. This makes it especially useful for building chatbots, retrieval-augmented generation (RAG) systems, and AI agents that can perform multi-step reasoning tasks. By abstracting much of the integration complexity, **LangChain** helps researchers and engineers focus more on application logic and less on low-level orchestration.

<img src="https://github.com/HandsOnPythonProgramming/Hands-On-Python-Programming/blob/main/Images/langchain_diagram.jpg?raw=true" width=840 />

**LangChain**, **LangGraph**, and **LangSmith** together form an integrated ecosystem for building, orchestrating, and monitoring applications powered by large language models. LangChain is responsible for providing the core building blocks to develop LLM applications, such as prompt chaining, tool integration, memory handling, and connecting models to external data sources. LangGraph extends this by enabling structured, stateful, and graph-based control flow for AI agents, making it possible to design complex multi-step reasoning systems with branching, loops, and long-running processes. LangSmith complements both by offering observability, debugging, and evaluation tools that let developers trace model behavior, inspect intermediate steps, test prompts, and measure performance in real-world usage. Together, they cover the full lifecycle of LLM applications—from building and orchestration to monitoring and optimization.

Note

---
The pipeline can be described as a sequential LLM workflow that moves through five core stages: retrieve, build prompt, call model, parse output, and next step.

First, in the retrieve stage, the system gathers relevant context from external sources such as vector databases, documents, APIs, or memory—this is where tools like retrieval-augmented generation (RAG) are typically used to enrich the input with domain-specific information. Next, in build prompt, the retrieved context is combined with the user query and system instructions to construct a structured prompt that gives the model clear guidance on what to do. Then, during call model, the prompt is sent to the LLM (via frameworks like LangChain), which processes the input and generates a raw response. In the parse output step, the system interprets the model’s response—extracting structured data, deciding whether tools should be invoked, or formatting the result into a usable schema. Finally, in next step, the system determines whether the workflow is complete or if further actions are required; if more reasoning, tool use, or branching is needed, control can be passed to a graph-based orchestrator like LangGraph, while execution traces and evaluation metrics are recorded in LangSmith for debugging and optimization.

# 12.3 Ollama

---

Description

---
**Ollama** is an open-source platform that simplifies running **large language models** (LLMs) locally on a personal computer, workstation, or cloud-based environment such as **Google Colab**. Instead of relying on external **AI services**, **Ollama allows users to download, manage, and execute models directly on their own hardware**. **Popular models supported by Ollama include LLaMA, Mistral, Gemma, DeepSeek, and many other open-source LLMs.**

One of Ollama's key advantages is its ease of use. A model can be downloaded and launched with a single command, eliminating much of the complexity typically associated with configuring and deploying large language models. Once running, Ollama starts a local HTTP server that exposes an API endpoint. **Applications and frameworks such as LangChain can communicate with this server by sending prompts and receiving generated responses, making it easy to integrate local AI models into custom workflows and applications.**

Because the models run locally, users maintain full control over their data and do not need to send prompts to external cloud providers. This can improve privacy, reduce operating costs, and **eliminate the need for API keys or internet connectivity during inference**. Ollama also handles model storage, loading, and resource management, providing a user-friendly interface for working with advanced AI models.

# 12.4  Example

---

Description

---
**Large language models (LLMs)** have transformed the field of artificial intelligence by demonstrating an impressive ability to understand natural language, generate human-like text, answer questions, and assist with a wide range of tasks. Despite their capabilities, however, they share a fundamental limitation: they can only generate responses based on the information available in their training data. Consequently, they have no inherent knowledge of recent events, private documents, organizational data, course materials, or any other information that was not included during training. When asked about such topics, an **LLM** may produce incomplete, outdated, or even fabricated responses because it lacks access to the underlying source material.

**Retrieval-Augmented Generation (RAG)** was developed to overcome this limitation by combining the reasoning capabilities of large language models with the ability to retrieve information from external knowledge sources. Rather than relying solely on information learned during training, a RAG system first searches a collection of user-provided documents—such as web pages, PDFs, research papers, notes, or textbooks—to identify information relevant to a user's question. The retrieved content is then supplied to the language model as additional context, allowing it to generate responses that are grounded in actual source material. This approach significantly improves accuracy, relevance, transparency, and reliability.

This workflow demonstrates the construction of a complete **RAG** system using the **LangChain framework** and a locally hosted **LLaMA 3 model** running through Ollama. The process begins with the installation and configuration of **Ollama**, followed by downloading the **LLaMA 3 model**, which provides the **system's natural language understanding, reasoning, and text-generation capabilities**. Once the language model is available, the source data that will serve as the **knowledge base** is loaded into the system. In this example, the source document is **Alice's Adventures in Wonderland**, although the same workflow can be applied to virtually any collection of documents.

After the source material is loaded, the text is divided into smaller, manageable segments known as chunks. These chunks are then transformed into numerical vector representations, called **embeddings**, which capture the semantic meaning of the text. The resulting embeddings are stored in a **FAISS vector database**, enabling efficient semantic search. Unlike traditional keyword-based search methods, semantic search identifies relevant passages based on meaning and context, making it possible to retrieve information even when the wording of a query differs from the wording used in the source document.

When a user submits a question, a retriever searches the vector database and identifies the text segments most relevant to the query. **LangChain then orchestrates** the interaction between the **retriever and the LLaMA 3 model** by passing the retrieved passages, along with the user's question, through a processing pipeline. The language model uses this retrieved context to formulate its response, ensuring that the generated answer is grounded in the source material rather than relying exclusively on its pre-trained knowledge.

By combining document retrieval with language generation, the resulting **RAG system can answer questions about information that was never part of the model's original training data**. This approach enables the creation of intelligent applications that provide accurate, context-aware, and source-grounded responses while leveraging the power of modern large language models.

Note

---
Alice in Wonderland will be used for this LangChain example. Once fetched using a Web loader, it becomes a LangChain Document object that can then be processed.

```python
# 6. Load webpage
from langchain_community.document_loaders import WebBaseLoader

url = "https://www.gutenberg.org/cache/epub/11/pg11-images.html"
loader = WebBaseLoader(url)
raw_documents = loader.load()
```

Note

---

The long document is broken down into smaller overlapping pieces using a text splitter. Chunking improves the retrieval accuracy because vector search works best on small, coherent sections of text.

```python
# 7. Split into chunks
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
)

documents = text_splitter.split_documents(raw_documents)
```

Note

---
This converts each chunk into a numerical vector using a HuggingFace embedding model. These vectors represent the meaning of each chunk so they can be searched semantically.

```python
# 8. Embeddings
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

Note

---
Each embedding is stored in an FAISS index, which is a "fast similarity-search database". This allows you to retrieve the most relevant chunks for any question.

```python
# 9. Vectorstore
from langchain_community.vectorstores import FAISS

vectorstore = FAISS.from_documents(documents, embeddings)
retriever = vectorstore.as_retriever()
```

Note

---
This is where it all comes together. The pipeline retrieves context, inserts it into a prompt, sends it to Ollama and generates the output.

```python
# 10. Connect Ollama to LangChain
from langchain_community.llms import Ollama
from langchain_core.runnables import RunnableParallel, RunnablePassthrough
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

template = """Use the following pieces of context to answer the user's question.
If you don't know the answer, say "I don't know."

Context:
{context}

Question:
{question}

Answer:
"""

prompt = PromptTemplate(
    template=template,
    input_variables=["context", "question"],
)

# Build the RAG chain
rag_chain = (
    {"context": retriever, "question": RunnablePassthrough()}
    | prompt
    | Ollama(model="llama3")
    | StrOutputParser()
)
```

```python
# 11. Provide query
query = "Who is the White Rabbit?"
answer = rag_chain.invoke(query)

print(answer)
```

```python
# 12. Example question that it cannot answer

query = "What is the fastest man-made object ever made?"
answer = rag_chain.invoke(query)

print(answer)
```
